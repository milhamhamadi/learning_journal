Kombinasi GroupBy dan querry dan aggregation

```python
df[
    (df['year_id']> 2010) &
    (df['fran_id']=="Spurs")
].groupby(['year_id','game_result'])['game_id'].count()
```

```python
df[
    (df['year_id']> 2010) &
    (df['fran_id']=="Spurs")
].groupby(['year_id','game_result']).agg({'pts' : 'mean','game_id' :'count'})
```

## Manipulating Columns

-create new column based on existing data
-rename columns
-drop columns

```python
df.sample(5)
```

```python
nba=df.copy()
#agar tidak merubah data asal
```

```python
#create new column based on existing
nba['selisih_poin'] = nba['pts'] - nba['opp_pts']
```

```python
nba.head()
```

```python
#rename column
nba.rename(columns={'game_location': 'location', 'game_result' : 'result'})
```

```python
nba.head()
#belum ada perubahan di nama column karena belum di save dan sebelum ini hanya preview
```

```python
#inplace untuk save
nba.rename(columns={'game_location': 'location', 'game_result' : 'result'}, inplace=True)
```

```python
nba.head(5)
```

```python
#drop columns
#untuk buang columns
nba.drop(['elo_i', 'elo_n'], axis=1, inplace=True)
```
